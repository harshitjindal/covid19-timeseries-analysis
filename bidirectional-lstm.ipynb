```python
from numpy import array
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Bidirectional
from sklearn.metrics import mean_squared_error
from sklearn import metrics 
import pandas as pd
```

```python
df = pd.read_excel('owid-covid-data .xlsx')
```

```python
df
```

```python
df = df[['location','total_cases']]
```

```python
df
```

```python
lis1 = dict()
for rows in df.itertuples():
    if rows.location not in lis1:
        lis1[rows.location] = []
    lis1[rows.location].append(rows.total_cases)
```

```python
X = []
y = []
for key in lis1.keys():
    lis = lis1[key]
    n = len(lis)
    for i in range(0,n-3):
        temp = lis[i:i+3]
        X.append(temp)
        y.append(lis[i+3])
```

```python
X = array(X)
X = X.reshape(len(X),1,3)
print(len(y))
X
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(100, activation='relu'), input_shape=(1, 3)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=300, verbose=2)
```
